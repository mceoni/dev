```python
import pyautogui
import time

def get_cursor_position():
    print("Coloque o cursor em um ponto da tela para capturar a posição. Pressione Ctrl+C para sair.")
    try:
        while True:
            # Obter posição atual do cursor
            x, y = pyautogui.position()
            print(f"Posição atual do cursor: X={x}, Y={y}", end="\r")
            time.sleep(0.1)  # Pequena pausa para evitar sobrecarga
    except KeyboardInterrupt:
        print("\nPrograma encerrado.")

# Executar a função
get_cursor_position()
```

```python
import pyautogui
import keyboard  # Biblioteca para detectar a tecla 'q'

def toggle_mouse_positions(p1, p2):
    while True:
        try:
            # Mover para a posição p1, clicar e digitar espaço
            pyautogui.moveTo(p1[0], p1[1])
            pyautogui.click()
            pyautogui.typewrite(" ")

            # Aguardar 60 segundos
            time.sleep(60)

            # Mover para a posição p2, clicar e digitar espaço
            pyautogui.moveTo(p2[0], p2[1])
            pyautogui.click()
            pyautogui.typewrite(" ")

            # Aguardar 180 segundos
            time.sleep(180)

            # Verificar se a tecla 'q' foi pressionada
            if keyboard.is_pressed('q'):
                print("Encerrando o programa...")
                break
        except KeyboardInterrupt:
            print("Programa interrompido manualmente.")
            break

# Coordenadas de exemplo para as posições p1 e p2
p1 = (500, 4000)  # Substitua pelas coordenadas desejadas
p2 = (2700, 600) # Substitua pelas coordenadas desejadas

# Executar o loop de alternância
toggle_mouse_positions(p1, p2)
```
